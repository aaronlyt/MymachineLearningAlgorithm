```python
class BPNN:
    """
    backpropgation neural network
    """
    def __init__(self,layer_nums,epochs,mini_batch_size,eta,w=None,b=None):
        """
        """
        self.layer_nums=layer_nums
        if w is None:
            self.w=[np.random.randn(x,y) for x,y in zip(layer_nums[1:],layer_nums[:-1])]
        else:
            self.w=w
        if b is None:
            self.b=[np.random.randn(x,1) for x in layer_nums[1:]]
        else:
            self.b=b
        
        self.epochs=epochs
        self.mini_batch_size=mini_batch_size
        self.eta=eta
        return None
    
    def feedForward(self,x):
        """
        return the output of each in the network if x is a input
        """
        self.z=[]
        self.a=[]
        self.a.append(x)
        for b,w in zip(self.b,self.w):
            t=np.dot(w,self.a[-1])+b
            self.z.append(t)
            self.a.append(1.0/(1+exp(-t)))
        self.a=np.array(self.a)
        self.z=np.array(self.z)
        return None
    
    def sigmoid(self,z):
        """
        the sigmoid function
        """
        return 1.0/(1+np.exp(-z))
    def sigmoid_derivative(self,z):
        """
        the derivative of sigmoid functioin
        """
        return self.sigmoid(z)*(1-self.sigmoid(z))
    
    def backPropgation(self,x,y):
        """
        backpropgate the error
        return a tuple (gradient_b,gradient_w)
        """
        self.feedForward(x)
        
        
        gradient_b=[np.zeros(b.shape) for b in self.b]
        gradient_w=[np.zeros(w.shape) for w in self.w]
        
        delta=[]
        delta= (self.a[-1]-y)*self.sigmoid_derivative(self.z[-1]) 
        
        #print self.a[-2].shape,delta.shape
        
        gradient_b[-1]=delta
        gradient_w[-1]=np.dot(delta,self.a[-2].transpose())
        
        for layer in np.arange(2,len(self.layer_nums)):
            
            z=self.z[-layer]
            sd=self.sigmoid_derivative(z)
            
            delta=np.dot(self.w[-layer+1].transpose(),delta)*sd
            
            gradient_b[-layer]=delta
            gradient_w[-layer]=np.dot(delta,self.a[-layer-1].transpose())
            
        return (gradient_b,gradient_w)
        
    def update_mini_batch(self,mini_batch):
        """
        update the weights and biases by applying gradient descent using backpropgation to a single mini batch
        """ 
        gradient_w_sum=[np.zeros(w.shape) for w in self.w]
        gradient_b_sum=[np.zeros(b.shape) for b in self.b]
        
      
        for x,y in mini_batch:
              
            gradient_b,gradient_w=self.backPropgation(x,y)
            
            #for nb,dnb in zip(gradient_w_sum, gradient_w):
                    #print nb.shape,dnb.shape
                    
            gradient_b_sum=[nb+dnb for nb, dnb in zip(gradient_b_sum, gradient_b)]
            
            gradient_w_sum=[nw+dnw for nw, dnw in zip(gradient_w_sum, gradient_w)]
            
        self.w=[w-(self.eta/len(mini_batch))*nw
                        for w, nw in zip(self.w, gradient_w_sum)]
        self.b=[b-(self.eta/len(mini_batch))*nb 
                       for b, nb in zip(self.b, gradient_b_sum)]
            
        return None
        
    def train(self,train_data):
        """
        train neural network using batch stochastic gradient descent
            
        """
        n=len(train_data)
            
        for j in np.arange(self.epochs):
                
            random.shuffle(train_data)
                
            mini_batches=[ train_data[k:k+self.mini_batch_size] 
                            for k in np.arange(0,n,self.mini_batch_size)
                            ]
            for mini_batch in mini_batches:
                self.update_mini_batch(mini_batch)
        return None
        
    def predict(self,test_X):
        """
        classification:
        predict the lable of each test sample
        """
        result=[]
        for x in test_X:
            self.feedForward(x)
            result.append(np.argmax(self.a[-1]))
        
        return np.array(result)
```

```python
"""
mnist_loader
~~~~~~~~~~~~

A library to load the MNIST image data.  For details of the data
structures that are returned, see the doc strings for ``load_data``
and ``load_data_wrapper``.  In practice, ``load_data_wrapper`` is the
function usually called by our neural network code.
"""

#### Libraries
# Standard library
import cPickle
import gzip

# Third-party libraries
import numpy as np

def load_data():
    """Return the MNIST data as a tuple containing the training data,
    the validation data, and the test data.

    The ``training_data`` is returned as a tuple with two entries.
    The first entry contains the actual training images.  This is a
    numpy ndarray with 50,000 entries.  Each entry is, in turn, a
    numpy ndarray with 784 values, representing the 28 * 28 = 784
    pixels in a single MNIST image.

    The second entry in the ``training_data`` tuple is a numpy ndarray
    containing 50,000 entries.  Those entries are just the digit
    values (0...9) for the corresponding images contained in the first
    entry of the tuple.

    The ``validation_data`` and ``test_data`` are similar, except
    each contains only 10,000 images.

    This is a nice data format, but for use in neural networks it's
    helpful to modify the format of the ``training_data`` a little.
    That's done in the wrapper function ``load_data_wrapper()``, see
    below.
    """
    f = gzip.open('data/mnist.pkl.gz', 'rb')
    training_data, validation_data, test_data = cPickle.load(f)
    f.close()
    return (training_data, validation_data, test_data)

def load_data_wrapper():
    """Return a tuple containing ``(training_data, validation_data,
    test_data)``. Based on ``load_data``, but the format is more
    convenient for use in our implementation of neural networks.

    In particular, ``training_data`` is a list containing 50,000
    2-tuples ``(x, y)``.  ``x`` is a 784-dimensional numpy.ndarray
    containing the input image.  ``y`` is a 10-dimensional
    numpy.ndarray representing the unit vector corresponding to the
    correct digit for ``x``.

    ``validation_data`` and ``test_data`` are lists containing 10,000
    2-tuples ``(x, y)``.  In each case, ``x`` is a 784-dimensional
    numpy.ndarry containing the input image, and ``y`` is the
    corresponding classification, i.e., the digit values (integers)
    corresponding to ``x``.

    Obviously, this means we're using slightly different formats for
    the training data and the validation / test data.  These formats
    turn out to be the most convenient for use in our neural network
    code."""
    tr_d, va_d, te_d = load_data()
    training_inputs = [np.reshape(x, (784, 1)) for x in tr_d[0]]
    training_results = [vectorized_result(y) for y in tr_d[1]]
    training_data = zip(training_inputs, training_results)
    validation_inputs = [np.reshape(x, (784, 1)) for x in va_d[0]]
    validation_data = zip(validation_inputs, va_d[1])
    test_inputs = [np.reshape(x, (784, 1)) for x in te_d[0]]
    test_data = zip(test_inputs, te_d[1])
    return (training_data, validation_data, test_data)


def vectorized_result(j):
    """Return a 10-dimensional unit vector with a 1.0 in the jth
    position and zeroes elsewhere.  This is used to convert a digit
    (0...9) into a corresponding desired output from the neural
    network."""
    e = np.zeros((10, 1))
    e[j] = 1.0
    return e
```
