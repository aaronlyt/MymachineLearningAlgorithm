```python
class GMM:
    """
    guassian mixture model
    """
    def __init__(self,n_components,tol=1e-2,min_covar=1e-3,max_iter=100,n_init=1,covariance_type='diag'):
        """
        init the parameters of the guassian model
        """
        self.n_components=n_components
        self.tol=tol
        self.min_covar=min_covar
        self.max_iter=max_iter
        self.n_init=n_init
        self.weights=np.ones(self.n_components)/(self.n_components*1.0)
        self.converged=False
        self.convariance_type=covariance_type
        return None
    def distribute_covar_matrix_to_match_covariance_type(self,
        tied_cv, covariance_type, n_components):
        """
        Create all the covariance matrices from a given template
        #assume all the guassian distribution has the same corvariance matrix
        """
        covar=[]
        for j in range(self.n_components):
            if covariance_type == 'diag':
                covar.append(tied_cv)
            else:
                raise ValueError("covariance_type must be one of  diag" )
        #print np.array(covar).shape
        return np.array(covar)
    def logsumexp(self,arr, axis=0):
        """
        Computes the sum of arr assuming arr is in the log domain.

        Returns log(sum(exp(arr))) while minimizing the possibility of
        over/underflow.

        Examples
        --------

        >>> import numpy as np
        >>> from sklearn.utils.extmath import logsumexp
        >>> a = np.arange(10)
        >>> np.log(np.sum(np.exp(a)))
        9.4586297444267107
        >>> logsumexp(a)
        9.4586297444267107
        """
        arr = np.rollaxis(arr, axis)
        # Use the max to normalize, as with the log this is what accumulates the less errors 
        #very important if no vmax = arr.max(axis=0) there will come up many problems
        vmax = arr.max(axis=0)
    
        out = np.log(np.sum(np.exp(arr - vmax), axis=0))
        out += vmax
        return out
    def _log_multivariate_normal_density_full(self,X, means, covars, min_covar=1.e-7):
        """
        Log probability for full covariance matrices.
        this function comes from scikit-learn mixture.gmm
        """
        n_samples, n_dim = X.shape
        nmix = len(means)
        log_prob = np.empty((n_samples, nmix))
        for c, (mu, cv) in enumerate(zip(means, covars)):
            try:
                cv_chol = linalg.cholesky(cv)
            except linalg.LinAlgError:
                print 'Exception excute'
                # The model is most probably stuck in a component with too
                # few observations, we need to reinitialize this components
                cv_chol = linalg.cholesky(cv + min_covar * np.eye(n_dim))
            cv_log_det = 2 * np.sum(np.log(np.diagonal(cv_chol)))
            cv_sol = linalg.solve(cv_chol, (X - mu).T).T
            log_prob[:, c] = - .5 * (np.sum(cv_sol ** 2, axis=1) +
                                 n_dim * np.log(2 * np.pi) + cv_log_det)

        return log_prob
    def score_samples(self,X):
        """
        calculate the possibilities which the i example belongs to the j components
        Returns
        -------
        logprob : array_like, shape (n_samples,)
            Log probabilities of each data point in X.

        responsibilities : array_like, shape (n_samples, n_components)
            Posterior probabilities of each mixture component for each
            observation
        """
        n_sample,n_dim=X.shape
        logprob=np.zeros(n_sample)
        logprob.astype('float')
        lpr=np.zeros((n_sample,self.n_components))
        lpr.astype('float')
        responsibilities=np.zeros((n_sample,self.n_components))
        responsibilities.astype('float')
        #maybe somthing wrong here
        for i in range(n_sample):
            for j in range(self.n_components):
                #if np.linalg.det(self.covar[j,:])>0:
                covar_det=np.linalg.det(self.covar[j])
               # else:
                   # covar_det=1
                convar_matrix=np.matrix(self.covar[j]).I
                lpr[i][j]=-0.5*(n_dim*np.log(2*np.pi)+np.log(covar_det)+np.dot(X[i,:]-self.means[j,:],convar_matrix).dot(X[i,:]-self.means[j,:]) )+np.log(self.weights[j])
    
        #lpr = (self._log_multivariate_normal_density_full(X, self.means, self.covar)
                   #+ np.log(self.weights))
        logprob=self.logsumexp(lpr,axis=1)
        #print 'logprob : inf ',logprob[logprob==inf]
        responsibilities=np.exp(lpr-logprob[:,np.newaxis])
        #print 'score_sample function  lpr:' ,lpr,lpr.shape
        #print 'score_sample function  responsibilities:' ,responsibilities,responsibilities[responsibilities==inf],responsibilities.shape
        return logprob,responsibilities
    
    def _do_mstep(self,X, responsibilities, min_covar):
        """
        maximum step to update parameters
        """
        weights=np.sum(responsibilities,axis=0)
        weighted_X_sum=np.dot(responsibilities.T,X)#n_components*n_feature
        inverse_weights=1.0/(weights[:,np.newaxis]+10*EPS)
        
        self.weights=(weights/(weights.sum()+10*EPS)+EPS)
        #print '_do_mstep:  self.weights ',self.weights
        n_features = X.shape[1]
    
        self.means=np.zeros_like(self.means)
        newcor=np.zeros_like(self.covar)
        for j in range(self.n_components):
            
            for i in range(X.shape[0]):
                self.means[j,:]+=responsibilities[i,j]*X[i,:]/(1.0*np.sum(responsibilities[:,j])+10*EPS)
                
                #newcor[j]+=responsibilities[i,j]*np.outer(X[i,:]-self.means[j,:],X[i,:]-self.means[j,:]) using the equation update corvariance has come problem
                newcor[j]+=responsibilities[i,j]*np.outer(X[i,:],X[i,:])
            """
            the annotation below comes from scikit-learn mixture.gmm
            post = responsibilities[:, j]
            # Underflow Errors in doing post * X.T are  not important
            np.seterr(under='ignore')
            avg_cv = np.dot(post * X.T, X) / (post.sum() + 10 * EPS)
            mu = self.means[j][np.newaxis]
            newcor[j,:] = (avg_cv - np.dot(mu.T, mu) + min_covar * np.eye(n_features))"""
            mu=self.means[j,:][np.newaxis]
            newcor[j]=newcor[j]/(1.0*np.sum(responsibilities[:,j]))-np.dot(mu.T,mu)
        print '_do_mstep: self.covar ',newcor
        self.covar=newcor
        #print  '_do_mstep:self.means  ',self.means
        return None
    def fit(self,X):
        """
        fit the data
        """
        X=np.array(X,dtype=np.float)
        
        max_log_prob=-np.inf
        best_params = {'weights': np.array([]),
                        'means': np.array([]),
                        'covars': np.array([])}
        for init_num in np.arange(self.n_init):
            
            self.means=cluster.KMeans(n_clusters=self.n_components).fit(X).cluster_centers_
            self.means.astype('float')
            #print 'initial self.means :',self.means
            self.weights=np.ones(self.n_components)/(self.n_components*1.0)
            """
            Estimate a covariance matrix, given data.
            Covariance indicates the level to which two variables vary together. 
            If we examine N-dimensional samples, X = [x_1, x_2, ... x_N]^T, then the covariance matrix element C_{ij} is the covariance of 
            x_i and x_j. The element C_{ii} is the variance of x_i.
            rowvar : int, optional
            If rowvar is non-zero (default), then each row represents a variable, with observations in the columns.
            Otherwise, the relationship is transposed: each column represents a variable, while the rows contain observations.
            """
            cv = np.cov(X,rowvar=0) + self.min_covar * np.eye(X.shape[1])
            #print cv.shape
            #self.covar=self.distribute_covar_matrix_to_match_covariance_type(cv,self.convariance_type,self.n_components)
            #self.covar = np.zeros((self.n_components, X.shape[1], X.shape[1]))+cv
            #self.covar *= np.identity( X.shape[1])
            self.covar = np.tile(cv, (self.n_components, 1, 1))
             # EM algorithms
            log_likelihood = []
            # reset self.converged_ to False
            self.converged_ = False
            
            for i in range(self.max_iter):
                curr_log_likelihood,responsibilities=self.score_samples(X)
                log_likelihood.append(curr_log_likelihood.sum())
                
                if (i>0) and abs(log_likelihood[-1]-log_likelihood[-2])<self.tol:
                    self.converged=True
                    break
                self._do_mstep(X, responsibilities, self.min_covar)
                
            #print log_likelihood[-1]    
            
            if (log_likelihood[-1]>max_log_prob):
                max_log_prob=log_likelihood[-1]
                best_params = {'weights': self.weights,
                                   'means': self.means,
                                   'covars': self.covar}
        if self.max_iter:
            print 'parameters assignment'
            #print self.covar
            #print self.means 
            self.covar = best_params['covars']
            self.means = best_params['means']
            self.weights= best_params['weights']
        return self
    def predict(test_X):
        """
        predict the class label,return the label
        """
        logprob,responsibilities=self.score_samples(test_X)
        return responsibilities.argmax(axis=1)
    def predict_prob(test_X):
        """
        predict the class label,return the label
        """
        logprob,responsibilities=self.score_samples(test_X)
        return responsibilities
```
