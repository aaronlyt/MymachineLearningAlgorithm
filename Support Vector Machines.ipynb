```python
class SVC():
    """
    support vector machines for classification
    """
    def __init__(self,C=1.0,kernel='linear',gamma=0.0,max_iter=40,tol=0.001):
        self.C=C
        self.kernel=kernel
        self.gamma=gamma
        self.max_iter=max_iter
        self.tol=tol
        self.alphas=None
        self.b=0
        self.w=None
        self.errorCache=None
        self.y_uniques=None
        return None
    def calcError(self,X,y,i):
        """
        calculate the term Ei
        """
        #print self.alphas*y,'\n'
        #print np.dot(X,X[i,:]),'\n'
        ui=(self.alphas*y).dot(np.dot(X,X[i,:]))+self.b
        #print ui
        Ei=ui-y[i]
        return Ei
    def selectj(self,X,y,i,Ei):
        """
        given the first ai,find the second aj,making the absolute value of ei-ej bigggest
        """
        maxj=-1
        maxDeltaE=0
        Ej=0
        self.errorCache[i]=[i,Ei]
        validation_indices=nonzero(self.errorCache[:,0])[0]
        if len(validation_indices)>1:
            for k in validation_indices:
                if k==i:continue
                Ek=self.calcError(X,y,k)
                deltaE=abs(Ek-Ei)
                if(deltaE>maxDeltaE):
                    maxDeltaE=deltaE
                    maxj=k
                    Ej=Ek
        else:
            j=i
            while j==i:
                j=int(random.uniform(0,len(X)))
            Ej=self.calcError(X,y,j)
        return maxj,Ej
    def updateEk(self,X,y,k):
        """
        store the error
        """
        Ek=self.calcError(X,y,k)
        self.errorCache[k]=[1,Ek]
        return None
    def solveLagrangeMultipliers(self,X,y,i):
        """
        calculate the ai and aj, and b
        """
        Ei=self.calcError(X,y,i)
        
        if ((y[i]*Ei<-self.tol) and (self.alphas[i]<self.C)) or ((y[i]*Ei)>self.tol and (self.alphas[i]>0)):
            j,Ej=self.selectj(X,y,i,Ei)
            alphaI_old=self.alphas[i]
            alphaJ_old=self.alphas[j]
            
            if(y[i]!=y[j]):
                L=max(0,self.alphas[j]-self.alphas[i])
                H=min(self.C,self.C+self.alphas[j]-self.alphas[i])
            else:
                L=max(0,self.alphas[j]+self.alphas[i]-self.C)
                H=min(self.C,self.alphas[j]+self.alphas[i])
            if L==H:
                print "L==H"
                return 0
            eta=X[i,:].dot(X[i,:])+X[j,:].dot(X[j,:])-2.0*X[i,:].dot(X[j,:])
            if eta>0:
                self.alphas[j]+=y[j]*(Ei-Ej)/eta*1.0
                if self.alphas[j]<L:
                    self.alphas[j]=L
                elif self.alphas[j]>H:
                    self.alphas[j]=H
            else:#in practice can be handled by calcaludate the objective function value lobj and hobj
                print 'eta<0'
                return 0
            self.updateEk(X,y,j)
            if(abs(self.alphas[j]-alphaJ_old)<0.00001):
                print 'j the second a not moving enough'
                return 0
            self.alphas[i]+=y[i]*y[j]*(alphaJ_old-self.alphas[j])
            self.updateEk(X,y,i)
            
            b1=self.b-Ei-y[i]*(self.alphas[i]-alphaI_old)*X[i,:].dot(X[i,:])-y[j]*(self.alphas[j]-alphaJ_old)*X[i,:].dot(X[j,:])
            b2=self.b-Ej-y[i]*(self.alphas[i]-alphaI_old)*X[i,:].dot(X[j,:])-y[j]*(self.alphas[j]-alphaJ_old)*X[j,:].dot(X[j,:])
            
            if((self.alphas[i]>0) and (self.alphas[i]<self.C)):
                self.b=b1
            elif ((self.alphas[j]>0) and (self.alphas[j]<self.C)):
                self.b=b2
            else:
                self.b=(b1+b2)/2.0
        return 1
    def smoP(self,X,y):
        """
        the outer loop of the smo
        """
        iteration_num=0
        entireSet=True
        alphaPairChanged=0
        while (iteration_num<self.max_iter) and ((alphaPairChanged>0) or entireSet):
            alphaPairChanged=0
            if entireSet:
                for i in range(len(X)):
                    alphaPairChanged+=self.solveLagrangeMultipliers(X,y,i)
                    #print 'fullset ,iter:%d i:%d, pairs changed %d' %(iteration_num,i,alphaPairChanged)
                iteration_num+=1
            else:
                nonBoundIndices=nonzero((self.alphas>0)*(self.alphas<self.C))[0]
                for i in nonBoundIndices:
                    alphaPairChanged+=self.solveLagrangeMultipliers(X,y,i)
                    #print 'fullset ,iter:%d i:%d, pairs changed %d' %(iteration_num,i,alphaPairChanged)
                iteration_num+=1
            if entireSet:entireSet=False
            elif alphaPairChanged==0:entireSet=True
            #print 'iteration number :%d ' % iteration_num
        return self.alphas,self.b
    def fit(self,X,y):
        """
        fit the data
        """
        self.alphas=np.zeros(len(X))
        self.errorCache=np.zeros((len(X),2))
        self.w=np.zeros(X.shape[1])
        if len(np.unique(y))!=2:
            print 'there more than 2 class label'
            return
        self.y_uniques=sorted(np.unique(y))
        y[y==self.y_uniques[0]]=-1
        y[y==self.y_uniques[1]]=1
        
        self.smoP(X,y)
        for i in range(len(X)):
            self.w+=self.alphas[i]*y[i]*X[i,:]
        #print y,self.y_uniques,'\n'
        return self
    def predict(self,test_X):
        """
        predict the test sample,classification
        """
        predictions=[]
        for test_vec in test_X:
            if np.dot(self.w,test_vec)+self.b>0:
                predictions.append(self.y_uniques[1])
            else:
                predictions.append(self.y_uniques[0])
        return np.array(predictions)
    
```

```python
print(__doc__)

import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# we create 40 separable points
np.random.seed(0)
X = np.r_[np.random.randn(20, 2) - [2, 2], np.random.randn(20, 2) + [2, 2]]
Y = [0] * 20 + [1] * 20

# fit the model
clf = svm.SVC(kernel='linear')
clf.fit(X, Y)

# get the separating hyperplane
w = clf.coef_[0]
print clf.coef_[0],clf.intercept_[0]
a = -w[0] / w[1]
xx = np.linspace(-5, 5)
yy = a * xx - (clf.intercept_[0]) / w[1]

# plot the parallels to the separating hyperplane that pass through the
# support vectors
b = clf.support_vectors_[0]
yy_down = a * xx + (b[1] - a * b[0])
b = clf.support_vectors_[-1]
yy_up = a * xx + (b[1] - a * b[0])

# plot the line, the points, and the nearest vectors to the plane
plt.plot(xx, yy, 'k-')
plt.plot(xx, yy_down, 'k--')
plt.plot(xx, yy_up, 'k--')

plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1],
            s=80, facecolors='none')
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Paired)

plt.axis('tight')
plt.show()
```
